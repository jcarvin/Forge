```python
class Cell(object):
    
    def __init__(self, row, col, board):
        self.row = row
        self.col = col
        self.board = board
        self.value = " "
        self.next_value = " "
        self.running = False
        self.change = False
        self.neighbors = []
        
    
    def find_neighbors(self):
        if self.row == 0:
            if self.col == 0:
                self.neighbors.append(self.board.board[self.row + 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row][self.col + 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col])
                self.neighbors.append(self.board.board[self.row + 1][len(self.board.board[0]) - 1])
                
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col])
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col + 1])
                self.neighbors.append(self.board.board[len(self.board.board) - 1][len(self.board.board[0]) - 1])
                self.neighbors.append(self.board.board[self.row][len(self.board.board[0]) - 1])
                
            elif self.col == len(self.board.board[0]) - 1:
                self.neighbors.append(self.board.board[self.row][self.col - 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col])
                self.neighbors.append(self.board.board[self.row + 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row + 1][0])
                
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col])
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col - 1])
                self.neighbors.append(self.board.board[len(self.board.board) - 1][0])
                self.neighbors.append(self.board.board[self.row][0])
                
            else:
                self.neighbors.append(self.board.board[self.row][self.col - 1])
                self.neighbors.append(self.board.board[self.row][self.col + 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col])  
                
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col])   
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col + 1])  
                self.neighbors.append(self.board.board[len(self.board.board) - 1][self.col - 1])          
                
        elif self.row == len(self.board.board) - 1:
            if self.col == 0:
                self.neighbors.append(self.board.board[self.row - 1][self.col])
                self.neighbors.append(self.board.board[self.row - 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row - 1][len(self.board.board[0]) - 1])
                self.neighbors.append(self.board.board[self.row][self.col + 1])
                
                self.neighbors.append(self.board.board[0][self.col])
                self.neighbors.append(self.board.board[0][self.col + 1])
                self.neighbors.append(self.board.board[0][len(self.board.board[0]) - 1])
                self.neighbors.append(self.board.board[self.row][len(self.board.board[0]) - 1])
                
            elif self.col == len(self.board.board[0]) - 1:
                self.neighbors.append(self.board.board[self.row][self.col - 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col])
                self.neighbors.append(self.board.board[self.row - 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row - 1][0])
                
                self.neighbors.append(self.board.board[0][self.col])
                self.neighbors.append(self.board.board[0][self.col - 1])
                self.neighbors.append(self.board.board[0][0])
                self.neighbors.append(self.board.board[self.row][0])
            else:
                self.neighbors.append(self.board.board[self.row][self.col - 1])
                self.neighbors.append(self.board.board[self.row][self.col + 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col])
                
                self.neighbors.append(self.board.board[0][self.col - 1])
                self.neighbors.append(self.board.board[0][self.col + 1])
                self.neighbors.append(self.board.board[0][self.col])
                
        else:
            if self.col == 0:
                self.neighbors.append(self.board.board[self.row - 1][self.col])
                self.neighbors.append(self.board.board[self.row - 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row][self.col + 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col])
                
                self.neighbors.append(self.board.board[self.row - 1][len(self.board.board[0]) - 1])
                self.neighbors.append(self.board.board[self.row][len(self.board.board[0]) - 1])
                self.neighbors.append(self.board.board[self.row + 1][len(self.board.board[0]) - 1])
                
            elif self.col == len(self.board.board[0]) - 1:
                self.neighbors.append(self.board.board[self.row - 1][self.col])
                self.neighbors.append(self.board.board[self.row - 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row][self.col - 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col])
                
                self.neighbors.append(self.board.board[self.row - 1][0])
                self.neighbors.append(self.board.board[self.row][0])
                self.neighbors.append(self.board.board[self.row + 1][0])
                
            else:
                self.neighbors.append(self.board.board[self.row][self.col - 1])
                self.neighbors.append(self.board.board[self.row][self.col + 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row - 1][self.col])
                self.neighbors.append(self.board.board[self.row + 1][self.col - 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col + 1])
                self.neighbors.append(self.board.board[self.row + 1][self.col])
            
        
    def determine_value(self):
        living_neighbors = [n for n in self.neighbors if n.value == "O"]
        dead_neighbors = [n for n in self.neighbors if n.value != "O"]
        if self.value == "O":
            if len(living_neighbors) == 3 or len(living_neighbors) == 2:
                self.next_value = "O"
                self.change = False
            elif len(living_neighbors) < 2:
                self.next_value = " "
                self.change = True
            elif len(living_neighbors) > 3:
                self.next_value = " "
                self.change = True
        else:
            if len(living_neighbors) == 3:
                self.next_value = "O"
                self.change = True
            else:
                self.change = False

    def change_value(self):
        self.value = self.next_value
        
class Board(object):
    
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols
        self.board = []
        self.cells = []
        
        self.start_board()
        
        
    def start_board(self):
        for x in range(self.rows):
            self.board.append([])
            for y in range(self.cols):
                cell = Cell(x, y, self)
                self.cells.append(cell)
                self.board[x].append(cell)
        for x in self.cells:
            x.find_neighbors()
                
                
    def begin(self, generations, coords=None):
        if coords:
            self.set_init_value(coords)
            
        self.print_board()
        
        self.running = True
        it = 0
        
        while self.running == True:
            it += 1
            if it == generations:
                self.running = False
            time.sleep(.008)
            self.check_cells()
            for x in self.cells:
                x.change_value()
            self.print_board()
                
    def set_init_value(self, coords):
        for coord in coords:
            if coord[0] <= len(self.board)-1 and coord[1] <= len(self.board[0])-1:
                self.board[coord[0]][coord[1]].value = "O"
    
    def check_cells(self):
        change_list = []
        for x in self.cells:
            x.determine_value()
            if x.change == True:
                change_list.append(x)
        if not change_list:
            self.running = False
            
    def print_board(self):
        clear_output(wait=True)
        view = []
        for didx, row in enumerate(self.board):
            view.append([])
            for idx, col in enumerate(row):
                view[didx].append(col.value)
        for row in view:
            x = (" ".join(row))
            print(x)
```
